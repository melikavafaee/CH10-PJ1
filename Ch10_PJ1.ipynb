طراحی و تحلیل سیستم‌های مرتبه اول و دوم

import

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import lti, step, freqresp
```

سیستم مرتبه اول

```python
num1 = [1]
den1 = [1, 1]
system1 = lti(num1, den1)
```

سیستم مرتبه دوم

```python
num2 = [1]
den2 = [1, 2, 2]
system2 = lti(num2, den2)
```

تحلیل پاسخ زمانی

```python
t1, y1 = step(system1)
t2, y2 = step(system2)
```

نمودار

```python
plt.figure(figsize=(10, 6))
plt.subplot(2, 1, 1)
plt.plot(t1, y1, label='First Order System')
plt.plot(t2, y2, label='Second Order System')
plt.title('Time Response')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.legend()
plt.grid()
```

تحلیل پاسخ فرکانسی

```python
w1, h1 = freqresp(system1)
w2, h2 = freqresp(system2)
```

نمودار

```python
plt.subplot(2, 1, 2)
plt.plot(w1, 20 * np.log10(abs(h1)), label='First Order System')
plt.plot(w2, 20 * np.log10(abs(h2)), label='Second Order System')
plt.title('Frequency Response')
plt.xlabel('Frequency [rad/s]')
plt.ylabel('Amplitude [dB]')
plt.legend()
plt.grid()

plt.tight_layout()
```
